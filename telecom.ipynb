```python
import pandas as pd
import numpy as np
```

```python
rdinfro_df=pd.read_csv("Telco-Customer-Churn.csv")#loading the data
```

```python
rdinfro_df.describe() #explains statistical values of data
```

```python
rdinfro_df.isnull() #checks if any null values in data
```

```python
rdinfro_df.head(10)
```

```python
#drop the column
rdinfro_df.drop("customerID",axis=1,inplace=True)
```

```python
rdinfro_df
```

```python
# Convert to numeric (some blank rows will become NaN)
rdinfro_df['TotalCharges'] = pd.to_numeric(rdinfro_df['TotalCharges'], errors='coerce')
rdinfro_df['TotalCharges'].isnull().sum()
```

```python
#Drop the values with null
rdinfro_df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
rdinfro_df
```

```python
# Binary encoding for Yes/No columns
rdinfro_df.replace({'Yes': 1, 'No': 0}, inplace=True)

# Convert 'gender'
rdinfro_df['gender'] = rdinfro_df['gender'].map({'Male': 1, 'Female': 0})

# One-Hot encode multi-category columns
rdinfro_df = pd.get_dummies(rdinfro_df, columns=['InternetService', 'Contract', 'PaymentMethod'], drop_first=True)
```

```python
#Exploratory data anlysis
```

```python
rdinfro_df['SeniorCitizen'] = rdinfro_df['SeniorCitizen'].astype(int)
```

```python
cols_to_convert = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 
                   'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Replace Yes/No with 1/0
for col in cols_to_convert:
    rdinfro_df[col] = rdinfro_df[col].replace({'Yes': 1, 'No': 0, 'No internet service': 0})
```

```python
rdinfro_df['MultipleLines'] = rdinfro_df['MultipleLines'].replace({'Yes': 1, 'No': 0, 'No phone service': 0})
```

```python
rdinfro_df.info()
```

```python
rdinfro_df.isnull()
```

```python
#Customer Segmentation
```

```python
# Create tenure bins
bins = [0, 12, 24, 48, 60, 72]
labels = ['0-12', '13-24', '25-48', '49-60', '61-72']
rdinfro_df['tenure_group'] = pd.cut(rdinfro_df['tenure'], bins=bins, labels=labels)

# Create MonthlyCharges bins
rdinfro_df['charges_group'] = pd.cut(rdinfro_df['MonthlyCharges'], bins=[0, 30, 60, 90, 120], labels=['Low', 'Medium', 'High', 'Very High'])

# Print sample
rdinfro_df[['tenure', 'tenure_group', 'MonthlyCharges', 'charges_group']].head()
```

```python
# Churn rate by tenure group
print(rdinfro_df.groupby('tenure_group')['Churn'].mean())

# Churn rate by charge group
print(rdinfro_df.groupby('charges_group')['Churn'].mean())

# Churn by contract
print(rdinfro_df.groupby('Contract_Two year')['Churn'].mean())  # 0 = No, 1 = Yes
```
